```python
from tkinter import *
import math
import random
raiz=Tk()
miImagen=PhotoImage(file="galaxia.gif")
raiz.title("Paradoja de Fermi")
raiz.resizable(False,False)
raiz.config(bg="black")
#-------------PATALLA PRINCIPAL---------------------------
PantallaPrincipal=Frame() #frame
PantallaPrincipal.config(bg="black")
PantallaPrincipal.config(width="450",height="200")
PantallaPrincipal.pack() #empaqueta el Frame 
#------------------FIN------------------------------------------
ImagenDeFondo=Label(PantallaPrincipal,image=miImagen)
ImagenDeFondo.place(x=0, y=0, relwidth=1, relheight=1)

#-----------------------CUADRO DE TEXTO---------------------------------------------------------------
CuadroDeTextoInicial=Label(PantallaPrincipal , text="Explorando la Paradoja de Fermi") #comienza el label
CuadroDeTextoInicial.config(font=("Comic Sans MS",16))
CuadroDeTextoInicial.config(fg="yellow")
CuadroDeTextoInicial.config(bg="DarkOrchid4")
CuadroDeTextoInicial.config(compound="center") 
CuadroDeTextoInicial.place(x=70,y=60) #empaqueta a Label sin afectar el tamaño
#--------------------------FIN--------------------------------------------------------

#--------------BOTON---------------
def codigoBoton():
    #radio=float(cuadroDeEntrada.get()) #utilizar "radio" para calculos en la probabilidad, etc.
    #textoFinal.config(text="El radio ingresado es " + str(radio))

    # Crear primera ventana
    ventana1 = Toplevel(raiz)
    ventana1.title("Via Lactea")
    ventana1.resizable(True,True)
    ventana1.config(bg="black")
    #nuestro lienzo(galaxia)
    galaxia = Canvas(ventana1, width=800,height=800,bg="black")
    galaxia.pack()
    #width=5
    xc = 390
    yc = 420
    xc2 = 390
    yc2 = 380
    #parametros de la espiral
    angulo = 0
    a = 6.5  #tamaño de la espiral 1.8
    b = 1.5 #distancia en vueltas 1.9
    for i in range (190):
        angulo_radianes = math.radians(angulo)
        r = a + b *angulo
        x = xc + r* math.cos(-1*angulo_radianes)
        y = yc + r* math.sin(-1*angulo_radianes)
        galaxia.create_oval(x-1, y-1, x+1, y+1, fill="white", outline = "slate gray",width=7)
        galaxia.create_oval(x-1, y-1, x+1, y+1, fill="white", outline = "light sky blue",width=3)#arriba 1
        galaxia.create_oval(x-1, y-1, x+1, y+1, fill="white", outline = "sky blue",width=2)
        
        x2 = xc2 + r* math.cos(-1*angulo_radianes+math.pi)
        y2 = yc2 + r* math.sin(-1*angulo_radianes+math.pi)
        galaxia.create_oval(x2-1, y2-1, x2 + 1, y2 + 1,fill="white",outline="slate gray",width=7)
        galaxia.create_oval(x2-1, y2-1, x2 + 1, y2 + 1,fill="white",outline="light sky blue",width=3)#abajo 1
        galaxia.create_oval(x2-1, y2-1, x2 + 1, y2 + 1,fill="white",outline="sky blue",width=2)
        

        angulo += 1

    xc =390 #460
    yc =425 #480
    xc2=380  #360
    yc2=375  #315
    angulo = 0
    a = 5 # tamaño de la espiral 1.5
    b = 2.5#distancia en vueltas 3.2

    for i in range(130):#130
        angulo_radianes = math.radians(angulo)
        r = a + b * angulo
        x = xc + r* math.cos(-1*angulo_radianes)
        y = yc + r* math.sin(-1*angulo_radianes)
        galaxia.create_oval(x-1, y-1, x + 1, y + 1,fill="white",outline="slate gray",width=7) #slate gray
        galaxia.create_oval(x-1, y-1, x + 1, y + 1,fill="white",outline="light sky blue",width=3)
        galaxia.create_oval(x-1, y-1, x + 1, y + 1,fill="white",outline="sky blue",width=2)#arriba 2
        
    
        x2 = xc2 + r* math.cos(-1*angulo_radianes+math.pi)
        y2 = yc2 + r* math.sin(-1*angulo_radianes+math.pi) #abajo 2
        galaxia.create_oval(x2-1, y2-1, x2 + 1, y2 + 1,fill="white",outline="slate gray",width=7)
        galaxia.create_oval(x2-1, y2-1, x2 + 1, y2 + 1,fill="white",outline="light sky blue",width=3)
        galaxia.create_oval(x2-1, y2-1, x2 + 1, y2 + 1,fill="white",outline="sky blue",width=2)
    
        angulo += 1

    galaxia.grid()
    #galaxia.configure(scrollregion=(-400,-300,300,400))#enfocar

    px=600
    py=550
    #nuestro planeta azul
    galaxia.create_oval(px,py,px+8,py+8,fill='blue')

    #disco
    nx=390
    ny=400

    galaxia.create_oval(nx-30,ny-20,nx+30,ny+20,fill='orange',width=0)

    galaxia.create_oval(nx-25,ny-15,nx+25,ny+15,fill='gold',width=0)

    galaxia.create_oval(nx-18,ny-8,nx+18,ny+8,fill='yellow',width=0)

    galaxia.create_oval(nx-35,ny-25,nx+35,ny+25,outline='light steel blue',width=7)

    galaxia.create_oval(nx-40,ny-30,nx+40,ny+30,outline='powder blue',width=3)

    galaxia.create_oval(nx-30,ny-20,nx+30,ny+20,outline='light goldenrod',width=5)

    def burbuja(event):
        r=5
        xc=px-r
        yc=py-r
        galaxia.create_oval(xc,yc,xc+4*r,yc+4*r,outline='red')

    #vinculando funcion burbuja con el evento del clic
    galaxia.bind("<Button-1>",burbuja)

    planetas=300 #dependera de drake

    # Función para generar el brillo de una estrella
    def generador_de_brillo(num_puntos):
        puntos = [] #lista vacia llamada puntos
        for numros_p in range(num_puntos): 
            x = random.randint(10, 790)
            #Se genera un valor aleatorio para la coordenada x dentro del rango de 10 a 790
            y = random.randint(10, 790)
            #Se genera un valor aleatorio para la coordenada y dentro del rango de 10 a 790
            tamaño_punto = 0.5 # Tamaño constante del punto
            color = "white" # Color blanco
            estrella = galaxia.create_oval(x, y, x+tamaño_punto, y+tamaño_punto, fill=color)
            #Se crea un óvalo en el lienzo con las coordenadas x, y, x+tamaño_punto, y+tamaño_punto y se llena con el color especificado
            puntos.append(estrella)
            #Se agrega la referencia del objeto de la estrella a la lista puntos
        for estrella in puntos:
            flash_puntos(estrella)
            #llama a la funcion flash

    # Función para hacer destellar un punto
    def flash_puntos(estrella):
        x = random.randint(10, 790)
        y = random.randint(10, 790)
        galaxia.coords(estrella, x, y, x+2, y+2)  # Cambiar la posición del punto
        galaxia.itemconfig(estrella, fill="white")  # Cambiar el color del punto a blanco
        galaxia.after(random.randint(100, 1000), lambda: parpadeo_puntos(estrella)) # Este número representa generado por un randon de 100 a 1000 es el tiempo de retardo en milisegundos antes de que se ejecute la función 

    # Función para ocultar un punto
    def parpadeo_puntos(estrellas):
        galaxia.itemconfig(estrellas, fill="black") # Cambiar el color del punto a negro
        galaxia.after(random.randint(100, 1000), lambda: flash_puntos(estrellas))
        # Este número representa generado por un randon de 100 a 1000 es el tiempo de retardo en milisegundos antes de que se ejecute la función 

    # Iniciar el brillo de las estrellas
    num_estrellas = 450 #numero de estrellas a generarse
    puntos = [] #se crea una lista vacia 
    for bri_start in range(num_estrellas):
        x = random.randint(10, 790)
        #Se genera un valor aleatorio para la coordenada x dentro del rango de 10 a 790
        y = random.randint(10, 790)
           #Se genera un valor aleatorio para la coordenada y dentro del rango de 10 a 790
        tamaño_punto = 2 # Tamaño constante del punto
        color = "white" # Color blanco
        punto = galaxia.create_oval(x, y, x+tamaño_punto, y+tamaño_punto, fill=color)
        #Se crea un óvalo en el lienzo con las coordenadas x, y
        puntos.append(punto)
        #Se agrega la referencia del objeto de la estrella a la lista

    # Llamar a la función para dibujar la Vía Láctea
    generador_de_brillo(num_estrellas)

    # Ejecutar el bucle principal de la aplicación 
    ventana1.mainloop()


boton=Button(PantallaPrincipal , text=("!Comenzar") , command=codigoBoton)
boton.place(x=190,y=110)
#---------------FIN---------------

#-------------------------TEXTO FINAL-------------------------------
textoFinal=Label(PantallaPrincipal)
textoFinal.config(bg="midnightblue")
textoFinal.config(fg="yellow")
textoFinal.place(x=60,y=200)
#-------------------------FIN---------------------------------


raiz.mainloop()
```
