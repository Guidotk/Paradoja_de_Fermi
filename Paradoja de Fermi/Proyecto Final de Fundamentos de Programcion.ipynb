```python
from tkinter import *
import math
import random
raiz=Tk()
miImagen=PhotoImage(file="galaxia.gif")
raiz.title("Paradoja de Fermi")
raiz.resizable(False,False)
raiz.config(bg="black")
#-------------PATALLA PRINCIPAL---------------------------
PantallaPrincipal=Frame() #frame
PantallaPrincipal.config(bg="black")
PantallaPrincipal.config(width="450",height="200")
PantallaPrincipal.pack() #empaqueta el Frame 
#------------------FIN------------------------------------------
ImagenDeFondo=Label(PantallaPrincipal,image=miImagen)
ImagenDeFondo.place(x=0, y=0, relwidth=1, relheight=1)

#-----------------------CUADRO DE TEXTO---------------------------------------------------------------
CuadroDeTextoInicial=Label(PantallaPrincipal , text="Explorando la Paradoja de Fermi") #comienza el label
CuadroDeTextoInicial.config(font=("Comic Sans MS",16))
CuadroDeTextoInicial.config(fg="yellow")
CuadroDeTextoInicial.config(bg="DarkOrchid4")
CuadroDeTextoInicial.config(compound="center") 
CuadroDeTextoInicial.place(x=70,y=60) #empaqueta a Label sin afectar el tama√±o
#--------------------------FIN--------------------------------------------------------

#--------------BOTON---------------
def codigoBoton():
    #radio=float(cuadroDeEntrada.get()) #utilizar "radio" para calculos en la probabilidad, etc.
    #textoFinal.config(text="El radio ingresado es " + str(radio))

    # Crear primera ventana
    ventana1 = Toplevel(raiz)
    ventana1.title("Via Lactea")
    ventana1.resizable(True,True)
    ventana1.config(bg="black")
    #nuestro lienzo(galaxia)
    galaxia= Canvas(ventana1,width=1000, height=800, bg='black')
    galaxia.pack()
    for angulo in range(0, 1440 , 10):
        radio = 100 + angulo/3
        x = 500 + radio * math.cos(math.radians(angulo))
        y = 400 + radio * math.sin(math.radians(angulo))
        #esta espiral esta basada en la de diego chavez
        galaxia.create_oval(x, y, x+4, y+4, fill='white')  #puntos blancos 
    px=900
    py=450
    #nuestro planeta azul
    galaxia.create_oval(px,py,px+4,py+4,fill='blue')

    planetas=300 #dependera de drake

    for i in range(planetas):
        x = random.randint(0, 1000)
        y = random.randint(0, 800)
        # restringiendo para que los planetas no esten cercanos al nuestro
        if not (840<=x<=940 and 400<=y<=500):
            planeta = galaxia.create_oval(x, y, x + 2, y + 2, fill='green')
    
    def burbuja(event):
        r=4 #luego le metemos random
        xc=px-r
        yc=py-r
        galaxia.create_oval(xc , yc, xc+4*r, yc+4*r, outline="red")
    
    galaxia.bind("<Button-1>",burbuja)

boton=Button(PantallaPrincipal , text=("!Comenzar") , command=codigoBoton)
boton.place(x=190,y=110)
#---------------FIN---------------

#-------------------------TEXTO FINAL-------------------------------
textoFinal=Label(PantallaPrincipal)
textoFinal.config(bg="midnightblue")
textoFinal.config(fg="yellow")
textoFinal.place(x=60,y=200)
#-------------------------FIN---------------------------------


raiz.mainloop()
```
