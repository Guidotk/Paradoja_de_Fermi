###   _4.1_. Ecuacion de Drake
Cuntas sociedades aliengenas existen y son detectables? Esta famosa frmula nos da una idea. La Ecuacin de Drake 
                                            N=R*.fp.ne.fl.fi.fc.L

-         N   : El nmero de civilizaciones en la Va Lctea cuyas emisiones electromagnticas son detectables.

-         R * : La tasa de formacin de estrellas aptas para el desarrollo de vida inteligente (nmero por ao).

-         fp  : La fraccin de esas estrellas con sistemas planetarios.

-         ne  : El nmero de planetas, por sistema solar, con un ambiente adecuado para la vida.

-         fl  : La fraccin de planetas adecuados en los que realmente aparece la vida.

-         fi  : La fraccin de planetas portadores de vida en los que emerge vida inteligente.

-         fc  : La fraccin de civilizaciones que desarrollan una tecnologa que produce signos detectables de su existencia.

-         L   : El promedio de tiempo que tales civilizaciones producen tales signos (aos).

Las estimaciones obtenidas por rigurosa investigacion:

- R* = 10
- fp = 0.05
- ne = 1
- fl = 0.8
- fi = 1
- fc = 0.1
- L = 1000000  
N = 10x0.05x1x0.8x1x0.1x1000000 = 40000

### 4.2.Distribucion Binomial
En la distribucin binomial, el resultado de cada ensayo se representa por una variable aleatoria X,representa el nmero de xitos en una serie de n ensayos independientes, donde cada ensayo solo tiene dos resultados posibles: xito o fracaso.La funcin de masa de probabilidad de la distribucin binomial est dada por:

....................................P(X = k) = (n choose k) * p^k * (1-p)^(n-k)...................................................

donde n es el nmero de ensayos, p es la probabilidad de xito de cada ensayo, k es el nmero de xitos y (n choose k) es el coeficiente binomial, que representa el nmero de formas en las que se pueden seleccionar k xitos de un total de n ensayos.

Donde : 

- n : el numero de civilizaciones que pueden emitir ondas 
- p : es la relacion de volumen de alcanze de las ondas entre el volumen total de la galaxia
- k : el numero de civilizaciones transmitoras a dectectar
- rt: radio de la burbuja
- rv : radio de la galaxia
- Pd : probanilidad de detectar k civilizaciones
por las revistas academicas tenemos que : 
n = 40000 obtenido por la ecuacion de drake
rt = 200
rv = 50000
p = (rt**3) / (Rv**3)
Pd = N * p * ((1 - p)**(N-1))
